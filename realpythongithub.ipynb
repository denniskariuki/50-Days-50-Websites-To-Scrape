```python
import csv
from bs4 import BeautifulSoup
import requests
import os




def check_url_status(url):
    try:
        page = requests.get(url)
        status_code = page.status_code
        if status_code==200:
            print(f'The Link {url} is accessible and the status Code is {status_code}')
            soup = BeautifulSoup(page.content, 'html.parser')
            
        else:
            print(f'The link {url} not accessibe and the status Code is {status_code}')
    except ConnectionError:
        print(f'Failled Connection to the {url}')


    
if __name__ =='__main__':
    link='https://realpython.github.io/fake-jobs/'
    check_url_status(link)
```

```python
url ='https://realpython.github.io/fake-jobs/'
page = requests.get(url)
soup1 = BeautifulSoup(page.content, 'html.parser')

def job_titles(soup1):
    job_titles_tags = soup1.find_all('h2', attrs={'class':'title is-5'})
    job_titles=[]
    for tags in job_titles_tags:
        job_titles.append(tags.text)
    return job_titles
```

```python
job_titles(soup1)
```

```python
def get_hiring_companies(soup1):
    hiring_companies_tags = soup1.find_all('h3', attrs={'class':'subtitle is-6 company'})
    hiring_companies=[]
    for tags in hiring_companies_tags:
        hiring_companies.append(tags.text)
    return hiring_companies
```

```python
get_hiring_companies(soup1)
```

```python
def get_date_time(soup1):
    date_time_tags = soup1.find_all('p',attrs={'class':'is-small has-text-grey'})
    dates =[]
    for tags in date_time_tags:
        dates.append(tags.text.strip())
    return dates
```

```python
get_date_time(soup1)
```
