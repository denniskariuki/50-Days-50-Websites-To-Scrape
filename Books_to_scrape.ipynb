```python
import csv
import pandas as pd
from bs4 import BeautifulSoup
import os
import requests
```

```python

def check_link_status(url):
    try:
        page = requests.get(url)
        status_code = page.status_code
        if status_code==200:
            print (f'The Link {url} is accessible, Status ={status_code} OK')
            soup = BeautifulSoup(page.content, 'html.parser')
            return soup
        else: 
            print(f'The Link {url} is not accessible , Status = {status_code}')
    except ConnectionError:
        print(f'Failed to connect to {url}')

#Scraping the Page Title:
        
def get_book_title(soup):
    book_title_tags =soup.find_all('h3')
    book_title=[]
    for tags in book_title_tags:
        book_title.append(tags.text)
    return book_title

# Scrape The Book Price..

def get_book_price(soup):
    book_price_tags= soup.find_all('p', attrs={'class':'price_color'})
    book_price =[]
    for tags in book_price_tags:
        book_price.append(tags.text)
    return book_price

# Scraping the Book Stars

def get_book_availability(soup):
    book_availability_tags = soup.find_all('p' , attrs={'class':'instock availability'})
    book_availability =[]
    for tags in book_availability_tags:
        book_availability.append(tags.text)
    return book_availability


#Iterating Through The Pages

def scrape_website(base_url, num_pages):
    all_titles = []
    all_price=[]
    all_availability=[]
    urls = [f"{base_url}/catalogue/page-{page}.html" for page in range(1, num_pages + 1)]
    for url in urls:
        print("Scraping page:", url)
        soup = check_link_status(url)
        if soup:
            titles = get_book_title(soup)
            prices = get_book_price(soup)
            availability=get_book_availability(soup)

        
            all_titles.extend(titles)
            all_price.extend(prices)
            all_availability.extend(availability)
          
    return all_titles

if __name__ == "__main__":

    link_to_check = "http://books.toscrape.com"
    check_link_status(link_to_check)

    base_url = "http://books.toscrape.com"
    num_pages = 50
    all_titles, all_price , all_availability = scrape_website(base_url, num_pages)
    
    df = pd.DataFrame({'Book Title': all_titles , 'Book Price': all_price , 'Book Availability': all_availability})
    df.to_csv('Books_to_scrape.csv', index=False)
 
```
