Kenya Moja Job Scraping Bot

```python
import csv
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd

#Checking the Status of the Link if it Scrapable 
def check_url(url):
    try:
        page = requests.get(url)
        status_code = page.status_code
        if status_code==200:
            print(f'The Link {url} is Accesssible , Status = {status_code} OK')
            soup = BeautifulSoup(page.content, 'html.parser')
            return soup
        else:
            print(f'The Link {url} is not Accessible , Status ={status_code}')
    except ConnectionError:
        print(f'Failed to Connect to {url} ')


url = 'https://www.kenyamoja.com/jobs'
page1 = requests.get(url)
soup1 = BeautifulSoup(page1.content, 'html.parser')

#Scraping the Job Title:
def get_job_title(soup1):
    job_title_tags =soup1.find_all('span', attrs ={'class':'items-title'})
    job_titles =[]
    for tags in job_title_tags:
        job_titles.append(tags.text)
    return job_titles

#Scraping the Exact Time The Job Was Posted:
def get_posted_date(soup1):
    posted_date_tags=soup1.find_all('time', attrs={'class':'datetime'})
    posted_dates=[]
    for tags in posted_date_tags:
        datetime_value = tags.get('datetime')
        date_and_time = datetime_value.split("T")[0] + " " + datetime_value.split("T")[1].split("Z")[0]
        posted_dates.append(date_and_time)
    return posted_dates

#Iterating Through all the pages 

def scrape_all_the_pages(base_url, num_of_pages):

    all_job_titles=[]
    all_posting_dates=[]

    urls =[f'{base_url}?page={page}' for page in range (1,num_of_pages + 1)]
    for url in urls:
        soup2 = check_url(url)
        print('Scraping ', url)
        
        if soup2:
            job_titles =get_job_title(soup1)
            posting_dates=get_posted_date(soup1)


            all_job_titles.extend(job_titles)
            all_posting_dates.extend(posting_dates)
        else:
            print(f'Cant Scrape at all')
        
    return all_job_titles, all_posting_dates



if __name__ == '__main__':
    link_to_check ='https://www.kenyamoja.com/jobs'
    check_url(link_to_check)

    base_url ='https://www.kenyamoja.com/jobs'
    num_of_pages =100

    all_job_titles , all_posting_dates = scrape_all_the_pages(base_url , num_of_pages)
    df = pd.DataFrame({'Job Title': all_job_titles, 'Posted On':all_posting_dates})
    df.to_csv('KenyaMoja.csv', index=False)

```
